```python
from pyspark.sql import SparkSession

# Load CSV from DBFS (or upload manually first)
df_bronze = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("/FileStore/machine_failure_data.csv")

# Save to Delta Lake as Bronze
df_bronze.write \
    .format("delta") \
    .mode("overwrite") \
    .save("/delta/bronze/machine_failure")
```
